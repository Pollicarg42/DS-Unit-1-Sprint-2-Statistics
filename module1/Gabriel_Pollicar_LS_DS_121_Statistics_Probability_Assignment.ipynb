# Apply the t-test to real data

Your assignment is to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

Your goals:

1. Load and clean the data (or determine the best method to drop observations when running tests)
2. Using hypothesis testing, find an issue that democrats support more than republicans with p < 0.01
3. Using hypothesis testing, find an issue that republicans support more than democrats with p < 0.01
4. Using hypothesis testing, find an issue where the difference between republicans and democrats has p > 0.1 (i.e. there may not be much of a difference)

Note that this data will involve *2 sample* t-tests, because you're comparing averages across two groups (republicans and democrats) rather than a single group against a null hypothesis.

```python
### YOUR CODE STARTS HERE

# Import tools and data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

link_congressData = "https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data"
df_congressData = pd.read_csv(link_congressData)

#Places the headers in appropriate columns
df_congressData.columns = ["Class-Name", "handicapped-infants",
"water-project-cost-sharing",
"adoption-of-the-budget-resolution",
"physician-fee-freeze",
"el-salvador-aid",
"religious-groups-in-schools",
"anti-satellite-test-ban",
"aid-to-nicaraguan-contras",
"mx-missile",
"immigration",
"synfuels-corporation-cutback",
"education-spending",
"superfund-right-to-sue",
"crime",
"duty-free-exports",
"export-administration-act-south-africa" ]

df_congressData.head(10)

```

```python
# Using hypothesis testing, find an issue that democrats support more than republicans with p < 0.01

# Creates subset of data with no "?" values
subset_congressData = df_congressData[~df_congressData['aid-to-nicaraguan-contras'].isin(null1)]
subset_congressDataNum = subset_congressData['aid-to-nicaraguan-contras'].eq('y').mul(1)
print(stats.ttest_1samp(subset_congressDataNum, .64))

# Creates subset of data with only republicans votes as binary
subset_republican = subset_congressData[(df_congressData['Class-Name'] == 'republican')]
## Turns y/n to 1 and 0
subset_republicanNum = subset_republican['aid-to-nicaraguan-contras'].eq('y').mul(1)


# Creates subset of data with only democrat votes as binary
subset_democrat = subset_congressData[(df_congressData['Class-Name'] == 'democrat')]
## Turns y/n to 1 and 0
subset_democratNum = subset_democrat['aid-to-nicaraguan-contras'].eq('y').mul(1)


# Two Sample T-test of both numbered republican subset and num. democrat subset
print(stats.ttest_ind(subset_republicanNum, subset_democratNum), "\n")


print("Democrats support this bill more than republicans(p-value = 6.83e-54", "\n")
# T-statistic is negative so democrats had higher average of supporting votes.
## p value is 6.835532456917403e-54
```

```python
# Using hypothesis testing, find an issue that republicans support more than democrats with p < 0.01

# Creates subset of data with no "?" values
subset_congressData = df_congressData[~df_congressData['religious-groups-in-schools'].isin(null1)]
subset_congressDataNum = subset_congressData['religious-groups-in-schools'].eq('y').mul(1)
print(stats.ttest_1samp(subset_congressDataNum, .64))

# Creates subset of data with only republicans votes as binary
subset_republican = subset_congressData[(df_congressData['Class-Name'] == 'republican')]
## Turns y/n to 1 and 0
subset_republicanNum = subset_republican['religious-groups-in-schools'].eq('y').mul(1)
# Creates subset of data with only democrat votes as binary
subset_democrat = subset_congressData[(df_congressData['Class-Name'] == 'democrat')]
## Turns y/n to 1 and 0
subset_democratNum = subset_democrat['religious-groups-in-schools'].eq('y').mul(1)
# Two Sample T-test of both numbered republican subset and num. democrat subset
print(stats.ttest_ind(subset_republicanNum, subset_democratNum), "\n")


print("republicans support this bill more than democrats(p-value = 3.41e-20 ", "\n")
# T-statistic is positive so republicans has higher average of supporting votes
## p value of 3.415e-20 
```

```python
# Using hypothesis testing, find an issue where the difference between republicans and democrats has p > 0.1

# Creates subset of data with no "?" values
subset_congressData = df_congressData[~df_congressData['water-project-cost-sharing'].isin(null1)]

print(stats.ttest_1samp(subset_congressDataNum, .5))

# Creates subset of data with only republicans votes as binary
subset_republican = subset_congressData[(df_congressData['Class-Name'] == 'republican')]
## Turns y/n to 1 and 0
subset_republicanNum = subset_republican['water-project-cost-sharing'].eq('y').mul(1)

# Creates subset of data with only democrat votes as binary
subset_democrat = subset_congressData[(df_congressData['Class-Name'] == 'democrat')]
## Turns y/n to 1 and 0
subset_democratNum = subset_democrat['water-project-cost-sharing'].eq('y').mul(1)

# Two Sample T-test of both numbered republican subset and num. democrat subset
print(stats.ttest_ind(subset_republicanNum, subset_democratNum), "\n")


print("Both republicans and democrats vote similar on this bill. p-value = 0.98", "\n")
# T-statistic is positive so republicans has higher average of supporting votes
## However p value is pvalue=0.9801332440116065 so no difference in
## republican or democrat support in this bill.
```
